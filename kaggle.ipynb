```python
def fill_value(df, col_list, fill=0):
    """
    a function to fill missing values
    @param df: dataframe
    @param col_list: list of column names
    @param fill: int (0) or str (mean, mode, median)
    return: preprocessed dataframe
    """
    if type(fill) == int or fill not in ["mean", "median", "mode"]:
        df[col_list] = df[col_list].fillna(fill)
    else:
        if fill == "mean":
            val = df[col_list].mean()
            val_map = val.to_dict()
        elif fill == "median":
            val = df[col_list].median()
            val_map = val.to_dict()
        else:
            val = df[col_list].mode()
            val_map = val.loc[0].to_dict()
        
        print(val_map)
        
        for col in col_list:
            df[col] = df[col].replace(np.nan, val_map[col])
        
    return df
```

```python
def fill_missing_value(df):   
  columns = df.columns
  df = fill_value(df, columns, fill="median")   
  return df
```

```python
train = fill_missing_value(train) 
```

```python
train_X = train[train.columns[1:-1]]
train_y = train[train.columns[-1]]
test_X = test[train.columns[1:-1]]
```

```python
random_model = RandomForestClassifier()
random_model.fit(train_X, train_y)
random_y_pred = random_model.predict(test_X)
```

```python
random_y_pred
```

```python
knn_model = KNeighborsClassifier()
knn_model.fit(train_X, train_y)
knn_y_pred = knn_model.predict(test_X)
knn_y_pred
```

```python
lr_model = LogisticRegression()
lr_model.fit(train_X, train_y)
lr_y_pred = lr_model.predict(test_X)
lr_y_pred
```

```python
pred = []
for i in range(len(random_y_pred)):
  vote_count = random_y_pred[i] + knn_y_pred[i] + lr_y_pred[i]
  if vote_count >= 2:
    pred.append(1)
  else:
    pred.append(0)
```

```python
res_df = pd.DataFrame({'id': range(len(pred)), 'label': pred})
```
